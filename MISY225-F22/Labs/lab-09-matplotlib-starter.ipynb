# Matplotlib

There are 8 basic functions as described at [http://mathonweb.com/help_ebook/html/functions_4.htm](http://mathonweb.com/help_ebook/html/functions_4.htm) 

Create a figure and do the following:

1. use the `seaborn` style sheet
2. create 4 subplots (2x2 2 rows with two subplots each row) with figure size (15, 15)
3. plot the following 3 types of functions: Linear function, Power function, Quadratic function (you can create the specific function yourself) and

- all functions can share the values of x, which should be 50 evenly spaced number between -10 and 10 (hint: use `np.linspace()`)
- add the function type as the title of the subplot
- the quadratic function has to be in the subplot in the first row and second column (you leave one subplot empty and arrange the other two anyway you w)
- the functions must have different line style, marker style, and color
- change the x label to x and y label to y
- add legends for each subplot to show the functions

A sample output is as follows:

![output](https://user-images.githubusercontent.com/595772/196346544-1f9a0582-8728-49fa-8fb7-b47db0a4850c.png)

```python
# import matplotlib and finish the figure
import matplotlib.pyplot as plt
import numpy as np


plt.style.use('seaborn')
fig, ax = plt.subplots(2, 2, figsize=(15, 15))

x = np.linspace(-10, 10, 50)
y1 = 1.8 * x
y2 = x**2 + 2*x + 2
y3 = x**3 + 0.5

ax[0, 0].plot(x, y1, label='y = 1.8x', linestyle='dashed', marker='o', color='goldenrod')
ax[0, 1].plot(x, y2, label='$y = x^2 + 2x + 2$', linestyle='dashdot', marker='s', color='green')
ax[1, 0].plot(x, y3, label='$y = x^3 + 0.5$', linestyle='solid', marker='*', color='coral')

ax[0, 0].set_xlabel('x')
ax[0, 1].set_xlabel('x')
ax[1, 0].set_xlabel('x')

ax[0, 0].set_ylabel('y')
ax[0, 1].set_ylabel('y')
ax[1, 0].set_ylabel('y')

ax[0, 0].set_title('Linear function')
ax[0, 1].set_title('Power function')
ax[1, 0].set_title('Quadratic function')

ax[0, 0].legend()
ax[0, 1].legend()
ax[1, 0].legend()
```
